```python
blood_sugar_men = [113,85,150,140,80,93,114,135,88, 77,82,129]
blood_sugar_women = [67,89,98,120,133,150,84,112,69,89,79,120,100]

plt.xlabel =['sugar range']
plt.ylabel = ['Total no of patients']
plt.title = ('Blood sugar Analysis')

plt.hist([blood_sugar_men,blood_sugar_women], bins = [80,100,125,200], rwidth = 0.4 ,label = ['MEN', 'WOMEN'],color = ['green',"orange"])

plt.legend()
plt.show()
```
